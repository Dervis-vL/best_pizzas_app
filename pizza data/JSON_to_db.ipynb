```python
import sqlite3
import json
```

```python
# create a database or conneect to existing one
conn = sqlite3.connect('IT_pizza_data.db')

# create cursor
c = conn.cursor()

# create table structure
c.execute("""CREATE TABLE pizza (
    rank integer,
    name text,
    city text,
    region text,
    address text,
    lat real,
    lon real
    )""")

# commit changes
conn.commit()

# close connection
conn.close()
```

```python
# create a database or conneect to existing one
conn = sqlite3.connect('IT_pizza_data.db')

# create cursor
c = conn.cursor()

pizzas = len(data.get('Italy'))
for pizza in range(pizzas):

    # submit into table
    c.execute("INSERT INTO pizza VALUES (:rank, :name, :city, :region, :address, :lat, :lon)",
            {
                'rank':data.get('Italy')[pizza].get('rank'),
                'name':data.get('Italy')[pizza].get('name'),
                'city':data.get('Italy')[pizza].get('city'),
                'region':data.get('Italy')[pizza].get('region'),
                'address':data.get('Italy')[pizza].get('address'),
                'lat':data.get('Italy')[pizza].get('geoloc').get('lat'),
                'lon':data.get('Italy')[pizza].get('geoloc').get('lng'),
            })

    # commit changes
    conn.commit()

# close connection
conn.close()
```
